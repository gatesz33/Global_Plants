```python
from PIL import Image #Importing Python Imaging Library used for open, convert, and edit images
import os #Importing library that allows interaction with file systems on local device or in environment

input_dir = r"C:\Users\gates\Downloads\power-plant-satellite-imagery-dataset-DatasetNinja\ds" #Defining file input data file path
output_dir = r"C:\Users\gates\Downloads\power-plant-satellite-imagery-dataset-DatasetNinja\ds_small" #Defining file output path (renaming to ds_small)
target_size = (224, 224) #Defining target size

count = 0

for root, dirs, files in os.walk(input_dir): #Recreating folder structure in output file path
    rel_path = os.path.relpath(root, input_dir)
    save_dir = os.path.join(output_dir, rel_path)
    os.makedirs(save_dir, exist_ok=True)

    for file in files: #Locating tif files and building file paths to the output
        if file.lower().endswith((".tif", ".tiff")): 
            input_path = os.path.join(root, file)

            
            output_name = os.path.splitext(file)[0] + ".jpg" #Saving output images as jpegs
            output_path = os.path.join(save_dir, output_name)

            try:
                with Image.open(input_path) as img: #Opening the images, converting, and resizing them as defined
                    img = img.convert("RGB")
                    img = img.resize(target_size)
                    img.save(output_path, "JPEG", quality=85)
                    count += 1
            except Exception as e:
                print("Skipped:", input_path, "| Error:", e) #Creating an outlet for any errors

print("Total images processed:", count) #Outputting total images processed to ensure all data was resized, converted, and saved in ds_small
```
